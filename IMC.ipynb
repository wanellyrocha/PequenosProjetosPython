Criado no s√©culo 19 pelo matem√°tico Lambert Qu√©telet, o √çndice de Massa Corporal, conhecido pela sigla IMC, √© um c√°lculo simples que permite medir se algu√©m est√° ou n√£o com o peso ideal. Ele aponta se o peso est√° adequado ou se est√° abaixo ou acima do peso.

Para fazer o c√°lculo, basta dividir o peso pela altura ao quadrado.

[IMC - Secretaria de Sa√∫de](https://www.saude.rj.gov.br/obesidade/noticias/2019/03/imc-voce-sabe-o-que-o-indice-de-massa-corporal-diz-sobre-sua-saude)

```python
altura = float(input('Digite sua altura em metros: '))
peso = float(input('Digite seu peso em KG: '))

imc = peso / altura**2
print('Seu IMC √© {:.2f}'.format(imc))

if imc < 16:
	print('Sua classifica√ß√£o IMC: Magreza grave')
elif imc < 17:
	print('Sua classifica√ß√£o IMC: Magreza moderada')
elif imc < 18.5:
	print('Sua classifica√ß√£o IMC: Magreza leve')
elif imc < 25:
	print('Sua classifica√ß√£o IMC: Saud√°vel')
elif imc < 30:
	print('Sua classifica√ß√£o IMC: Sobrepeso')
elif imc < 35:
	print('Sua classifica√ß√£o IMC: Obesidade Grau I')
elif imc < 40:
	print('Sua classifica√ß√£o IMC: Obesidade Grau II (severa)')
else:
	print('Sua classifica√ß√£o IMC: Obesidade Grau III (m√≥rbida)')
```
